# Functions Definition

### Baseline model with no constraints

```python
# Pulp baseline model - with no additional constraints (no carbon tax, no land limitation, no subsidy)
def baseline_optimisation_problem(TechnologiesDataFrame, YearsDataFrame):
    # Generate variables

    # Technologies related variables
    technologies = TechnologiesDataFrame.Technologies.tolist()
    initial_capacity = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Initial_Capacity_GW)) # in GW
    capacity_factor = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Capacity_Factor))
    efficiency = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Efficiency))
    capex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Capex_USDperGW)) # in $/GW
    variable_opex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Variable_Opex_USDperGWh)) # in $/GWh
    fixed_opex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Fixed_Opex_USDperGWhperyear)) # in $/GWh/year
    land_use = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Land_Use_km2perGW)) # in km2/GW
    emissions_factor = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Land_Use_km2perGW)) # in tCO2e/GWh_th
    
    # Yearly related variables
    years = YearsDataFrame.Year.tolist()
    demand = dict(zip(YearsDataFrame.Year, YearsDataFrame.Electricity_Demand_GWh)) # in GWh
    fossil_costs = dict(zip(YearsDataFrame.Year, YearsDataFrame.Fossil_Fuel_Costs_USDperGWh_th)) # in $/GWh_th
    
    # General variables defined out of spreadsheet
    operational_hours = 8760 # in hours/year
    discount = 0.1 # discount rate fraction

    #Initialise model
    model = pulp.LpProblem("Power_System", pulp.LpMinimize)
    
    # Decision variables
    new_capacity = pulp.LpVariable.dicts("New_Capacity", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GW
    cumulative_capacity = pulp.LpVariable.dicts("Cumulative_Capacity", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GW
    dispatch = pulp.LpVariable.dicts("Dispatch", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GWh
    
    # Optimisation objective
    model += pulp.lpSum(new_capacity[(t,y)]*capex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(dispatch[(t,y)]*variable_opex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(cumulative_capacity[(t,y)]*fixed_opex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(dispatch[('fossil',y)]/efficiency['fossil']*fossil_costs[y]/((1+discount)**(y-2020)) for y in years)\
    
    # Cumulative capacity calc (Indirect constraint)
    for t in technologies:
        for i in range(0,len(years)):
            if i==0:
                model += cumulative_capacity[(t,years[i])]==initial_capacity[t] + new_capacity[(t,years[i])]
            else:
                model += cumulative_capacity[(t,years[i])]==cumulative_capacity[(t,years[i-1])] + new_capacity[(t,years[i])]
    
    # Constraint 1: Supply = Demand
    for y in years: 
        model += pulp.lpSum(dispatch[(t,y)] for t in technologies) == demand[y]
    
    #Constraint 2: Max generation GWh <= Cumulative installed capacity * working hours
    for y in years: 
        for t in technologies: 
            model += dispatch[(t,y)] <= cumulative_capacity[(t,y)]*operational_hours*capacity_factor[t]
    
    # Solve model
    model.solve(pulp.PULP_CBC_CMD(msg=False))
    
    # Print out key model outputs
    print("Model status: ", pulp.LpStatus[model.status])
    print("Optimised net present value of the electrical system is: ", round(pulp.value(model.objective)/1000000,2),"m$",'\n')

    return_new_capacity = {(t,y): new_capacity[(t,y)].value() for t in technologies for y in years}
    return_cumulative_capacity = {(t,y): cumulative_capacity[(t,y)].value() for t in technologies for y in years}
    return_dispatch = {(t,y): dispatch[(t,y)].value() for t in technologies for y in years}

    return return_new_capacity, return_cumulative_capacity, return_dispatch, technologies, years
```

```python
# Carbon Tax Model
def carbon_tax_optimisation_problem(TechnologiesDataFrame, YearsDataFrame, carbon_tax):
    # Generate variables

    # Technologies related variables
    technologies = TechnologiesDataFrame.Technologies.tolist()
    initial_capacity = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Initial_Capacity_GW)) # in GW
    capacity_factor = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Capacity_Factor))
    efficiency = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Efficiency))
    capex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Capex_USDperGW)) # in $/GW
    variable_opex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Variable_Opex_USDperGWh)) # in $/GWh
    fixed_opex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Fixed_Opex_USDperGWhperyear)) # in $/GWh/year
    land_use = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Land_Use_km2perGW)) # in km2/GW
    emissions_factor = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Emissions_Factor_tCO2perGWh_th)) # in tCO2e/GWh_th
    
    # Yearly related variables
    years = YearsDataFrame.Year.tolist()
    demand = dict(zip(YearsDataFrame.Year, YearsDataFrame.Electricity_Demand_GWh)) # in GWh
    fossil_costs = dict(zip(YearsDataFrame.Year, YearsDataFrame.Fossil_Fuel_Costs_USDperGWh_th)) # in $/GWh_th
    
    # General variables defined out of spreadsheet
    operational_hours = 8760 # in hours/year
    discount = 0.1

    #Initialise model
    model2 = pulp.LpProblem("Power_System_with_carbon_tax", pulp.LpMinimize)
    
    # Decision variables
    new_capacity = pulp.LpVariable.dicts("New_Capacity_Carbon", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GW
    cumulative_capacity = pulp.LpVariable.dicts("Cumulative_Capacity_Carbon", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GW
    dispatch = pulp.LpVariable.dicts("Dispatch_Carbon", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GWh
    
    # Optimisation objective
    model2 += pulp.lpSum(new_capacity[(t,y)]*capex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(dispatch[(t,y)]*variable_opex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(cumulative_capacity[(t,y)]*fixed_opex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(dispatch[('fossil',y)]/efficiency['fossil']*fossil_costs[y]/((1+discount)**(y-2020)) for y in years) \
            + pulp.lpSum(dispatch[('fossil',y)]/efficiency['fossil']*emissions_factor['fossil']*carbon_tax/((1+discount)**(y-2020)) for y in years)
    
    # Cumulative capacity calc
    for t in technologies:
        for i in range(0,len(years)):
            if i==0:
                model2 += cumulative_capacity[(t,years[i])]==initial_capacity[t] + new_capacity[(t,years[i])]
            else:
                model2 += cumulative_capacity[(t,years[i])]==cumulative_capacity[(t,years[i-1])] + new_capacity[(t,years[i])]
    
    # Constraint 1: Supply = Demand
    for y in years: 
        model2 += pulp.lpSum(dispatch[(t,y)] for t in technologies) == demand[y]
    
    #Constraint 2: Max generation GWh <= Cumulative installed capacity * working hours
    for y in years: 
        for t in technologies: 
            model2 += dispatch[(t,y)] <= cumulative_capacity[(t,y)]*operational_hours*capacity_factor[t]
    
    # Solve model
    model2.solve(pulp.PULP_CBC_CMD(msg=False))
    
    # Print out key model outputs
    print("Model status: ", pulp.LpStatus[model2.status])
    print("Optimised net present value of the electrical system is: ", round(pulp.value(model2.objective)/1000000,2),"m$",'\n')
    
    return_new_capacity = {(t,y): new_capacity[(t,y)].value() for t in technologies for y in years}
    return_cumulative_capacity = {(t,y): cumulative_capacity[(t,y)].value() for t in technologies for y in years}
    return_dispatch = {(t,y): dispatch[(t,y)].value() for t in technologies for y in years}

    return return_new_capacity, return_cumulative_capacity, return_dispatch, technologies, years
```

```python
# Land Limitation Model
def land_optimisation_problem(TechnologiesDataFrame, YearsDataFrame, limited_area):
    # Generate variables

    # Technologies related variables
    technologies = TechnologiesDataFrame.Technologies.tolist()
    initial_capacity = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Initial_Capacity_GW)) # in GW
    capacity_factor = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Capacity_Factor))
    efficiency = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Efficiency))
    capex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Capex_USDperGW)) # in $/GW
    variable_opex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Variable_Opex_USDperGWh)) # in $/GWh
    fixed_opex = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Fixed_Opex_USDperGWhperyear)) # in $/GWh/year
    land_use = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Land_Use_km2perGW)) # in km2/GW
    emissions_factor = dict(zip(TechnologiesDataFrame.Technologies, TechnologiesDataFrame.Emissions_Factor_tCO2perGWh_th)) # in tCO2e/GWh_th
    
    # Yearly related variables
    years = YearsDataFrame.Year.tolist()
    demand = dict(zip(YearsDataFrame.Year, YearsDataFrame.Electricity_Demand_GWh)) # in GWh
    fossil_costs = dict(zip(YearsDataFrame.Year, YearsDataFrame.Fossil_Fuel_Costs_USDperGWh_th)) # in $/GWh_th
    
    # General variables defined out of spreadsheet
    operational_hours = 8760 # in hours/year
    discount = 0.1

    #Initialise model
    model3 = pulp.LpProblem("Power_System_with_ladn_limitation", pulp.LpMinimize)
    
    # Decision variables
    new_capacity = pulp.LpVariable.dicts("New_Capacity_Land", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GW
    cumulative_capacity = pulp.LpVariable.dicts("Cumulative_Capacity_Land", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GW
    dispatch = pulp.LpVariable.dicts("Dispatch_Land", [(t,y) for t in technologies for y in years], lowBound=0, cat="Continuous") # in GWh
    
    # Optimisation objective
    model3 += pulp.lpSum(new_capacity[(t,y)]*capex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(dispatch[(t,y)]*variable_opex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(cumulative_capacity[(t,y)]*fixed_opex[t]/((1+discount)**(y-2020)) for t in technologies for y in years) \
            + pulp.lpSum(dispatch[('fossil',y)]/efficiency['fossil']*fossil_costs[y]/((1+discount)**(y-2020)) for y in years)
    
    # Cumulative capacity calc
    for t in technologies:
        for i in range(0,len(years)):
            if i==0:
                model3 += cumulative_capacity[(t,years[i])]==initial_capacity[t] + new_capacity[(t,years[i])]
            else:
                model3 += cumulative_capacity[(t,years[i])]==cumulative_capacity[(t,years[i-1])] + new_capacity[(t,years[i])]
    
    # Constraint 1: Supply = Demand
    for y in years: 
        model3 += pulp.lpSum(dispatch[(t,y)] for t in technologies) == demand[y]
    
    #Constraint 2: Max generation GWh <= Cumulative installed capacity * working hours
    for y in years: 
        for t in technologies: 
            model3 += dispatch[(t,y)] <= cumulative_capacity[(t,y)]*operational_hours*capacity_factor[t]
    
    # Constraint 3: Land limitation
    model3 += (cumulative_capacity[('renewable', 2030)] - initial_capacity['renewable'])*land_use['renewable'] <= limited_area
    
    # Solve model
    model3.solve(pulp.PULP_CBC_CMD(msg=False))
    
    # Print out key model outputs
    print("Model status: ", pulp.LpStatus[model3.status])
    print("Optimised net present value of the electrical system is: ", round(pulp.value(model3.objective)/1000000,2),"m$",'\n')
    
    return_new_capacity = {(t,y): new_capacity[(t,y)].value() for t in technologies for y in years}
    return_cumulative_capacity = {(t,y): cumulative_capacity[(t,y)].value() for t in technologies for y in years}
    return_dispatch = {(t,y): dispatch[(t,y)].value() for t in technologies for y in years}

    return return_new_capacity, return_cumulative_capacity, return_dispatch, technologies, years
```
